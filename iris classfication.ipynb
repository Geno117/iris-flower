```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
import sys
```

```python
#f=open('iris.data','r')
names=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm','class']
frame = pd.read_csv('iris.data',names=names)
```

```python
#one hot classification
#try using tf onehot next time
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelBinarizer
Y_=frame['class']
X_=frame
X_=frame.drop(['class'],axis=1)
enc= LabelEncoder()
ohe= OneHotEncoder()
lbin= LabelBinarizer()
#print(f'before: \n {Y_}')
Y_=pd.DataFrame(lbin.fit_transform(Y_))
print(f'after: \n {Y_}')

#split training, dev,and test set using 80-10-10 split

from sklearn.model_selection import train_test_split

X_train, X_, Y_train, Y_ = train_test_split(X_, Y_, test_size=0.2,random_state=0)
X_dev, X_test, Y_dev, Y_test = train_test_split(X_,Y_,test_size=0.5,random_state=0)
```

```python
#checking sizes
print(X_train.shape)
print(Y_train.shape)
print(X_dev.shape)
print(Y_dev.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
#checking against frame
for i in range(5):
    sample_idx = np.random.choice(X_train.index)
    print(f'{X_train.loc[[sample_idx]]} : {Y_train.loc[sample_idx]}')
    print(f'{sample_idx} in frame \n{frame.loc[[sample_idx]]}')
    
```

```python
#creating a Neural Network
from keras.models import Model, load_model, Sequential
from keras.layers import Dense, Activation, Dropout, Input, Masking, TimeDistributed, LSTM, Conv1D
from keras.layers import GRU, Bidirectional, BatchNormalization, Reshape
from keras.optimizers import Adam
def model(input_shape):
    X_input=Input(shape=input_shape)
#4 dense layers of 4. Do the dense layers require activation?
    X=Dense(units=4)(X_input)
    X=Dense(units=4)(X)
    X=Dense(units=4)(X)
    X=Dense(units=3)(X)
#softmax activation
    act=Activation('softmax')(X)
    model=Model(inputs=[X_input],outputs=[act])
    return model
```

```python
X_train
X_train_np=X_train.to_numpy()
print(X_train_np.shape)
Y_train_np=Y_train.to_numpy()
print(Y_train_np.shape)
```

```python
myModel = model((4,))
myModel.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
myModel.summary()
```

```python
myModel.fit(X_train_np,Y_train_np,epochs=100,batch_size=30)
```

```python
X_dev_np=X_dev.to_numpy()
Y_dev_np=Y_dev.to_numpy()
```

```python
predictions = myModel.predict(X_dev)
```

```python
count=0
total=len(Y_dev)+1
for y_pred,y_truth in zip(predictions,Y_dev_np):
    y_pred_=np.argmax(y_pred)
    y_truth_=np.argmax(y_truth)
    print(f'{y_pred_}   {y_truth_}')
    if y_pred_==y_truth_:
        count+=1
accuracy=count/total
print(f'accuracy: {accuracy}')
```
